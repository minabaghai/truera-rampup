### Data Processing

```python
cancer_stats.columns
```

```python
# Drop uneccessary columns
cancer_stats.drop(columns=['id', 'Unnamed: 32', 'radius + texture mean'], axis=1, inplace=True)
```

```python
# Drop duplicate rows
cancer_stats.drop_duplicates(inplace=True)
```

```python
# How many missing values
cancer_stats.isna().sum()
```

```python
# Looking for outliers
sns.boxplot(x='area_se', data=cancer_stats);
```

```python
# Remove outliers
cancer_stats = cancer_stats[cancer_stats['area_se'] < 200]
```

### Feature Work

```python
# Separate into numerical and categorical
categorical = cancer_stats.select_dtypes([object, 'category'])
numerical = cancer_stats.select_dtypes(np.number)
```

```python
# One hot encoding
categorical = pd.get_dummies(categorical, drop_first = True, dtype='int64')
categorical.rename(columns={'diagnosis_M': 'Malignant Diagnosis'}, inplace=True)
categorical.head()
```

```python
# Standardize numerical variables

scaler = StandardScaler()
scaler.fit(numerical)
numerical_scaled = scaler.transform(numerical)
numerical = pd.DataFrame(numerical_scaled, columns=numerical.columns)
numerical.head()
```

```python
# Join numerical and categorical back
X = pd.concat([numerical, categorical.reset_index()], axis=1)
X.head()
```

### Splitting Data

```python
from sklearn.model_selection import train_test_split
```

```python
target = X['Malignant Diagnosis']
features = X.drop(columns=['Malignant Diagnosis', 'index'])

X_train, X_test, y_train, y_test = train_test_split(features, target, test_size = 0.2)
```

### Training a Model

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
model.fit(X=X_train, y=y_train)
```

### Get Model Predictions

```python
model.predict(X_test.values[:5])
```

```python
y_test.values[:5]
```

```python
predictions = model.predict(X_test)
X = X_test.reset_index().copy()
X['target'] = y_test.tolist()
X['prediction'] = predictions
X
```

```python
sns.regplot(x='target', y='prediction', data=X, logistic=True, ci=None);
```

```python
model.score(X_test, y_test)
```

```python
# Confusion matrix: [true positives, false positives]
#                   [false negatives, true negatives]

from sklearn.metrics import confusion_matrix

cf_matrix = confusion_matrix(y_test, predictions)
print(cf_matrix)
```

```python
# ROC/AUC Curve
import sklearn.metrics as metrics

fpr, tpr, threshold = metrics.roc_curve(y_test, predictions)
p = sns.lineplot(x=fpr, y=tpr)
p.set_title('ROC Curve')
```
